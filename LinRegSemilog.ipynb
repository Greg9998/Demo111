```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

x=[13.23,15.68,16.47,26.26,34.04];
w=[27.1428,27.2727,30.0653,34.3023,38.125];

x,w = np.array(x).reshape((-1, 1)), np.array(w)

linmodel = LinearRegression()
linmodel.fit(np.log10(x), w)

w_pred = linmodel.predict(np.log10(x))
w_20 = linmodel.predict(np.log10(np.array(20).reshape((-1, 1))))


plt.plot(x,w, 'o')
plt.plot(x,w_pred)
plt.plot(20,w_20, 'o')

plt.xscale('log')
plt.grid(which='both')
w_str=list(map('w={:.2f}%'.format,w_20))
plt.text(20, w_20, w_str)

ax=plt.gca()
xlab=np.arange(12, 40, 2).tolist()
plt.xticks(xlab, [ f"{x:.0f}" for x in xlab ])
```
